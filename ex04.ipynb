```python
import pandas as pd
data = pd.read_excel('./data/부산탄소발생현황.xlsx')
data
```

```python
data=data.rename(columns = {'배출량원단위(전기)':'전기','배출량원단위(가스)':'가스','배출량원단위(지역난방)':'지역난방'})
data.head()
```

```python
#연도별  배출량원단위(전기)	배출량원단위(가스)	배출량원단위(지역난방)

#group by 
dfg=data.groupby(['연도'])['전기','가스','지역난방'].mean()
dfg
```

```python
# 행열변환
dfg = dfg.T
```

```python
dfg
```

```python
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc
font_name = font_manager.FontProperties(fname="c:/Windows/Fonts/malgun.ttf").get_name()
rc('font', family=font_name)

dfg.plot(kind='bar',figsize=(10,6))
plt.show()
```

```python
#구별로 사용량 표시하기 

#구 위도경도 데이터 읽어오기
busan = pd.read_excel('./data/부산구청.xlsx')
busan.index
```

```python
#group by 
data.groupby(['연도','구명'])['전기','가스','지역난방'].sum()
```

```python
##2013 별도로 추출
df2013 = data[data['연도']==2013]
df2013.tail()
```

```python
#group by 
df_2013_gu = df2013.groupby(['구명'])['전기','가스','지역난방'].sum()
df_2013_gu
```

```python
import folium
map_osm = folium.Map(location=[35.166804,129.083479], zoom_start=12)   # 지도 초기 로딩위치
```

```python
for item in busan.index:
    lat = busan.loc[item,'위도']
    long = busan.loc[item,'경도']
    item = item.strip()
    folium.CircleMarker([lat,long],  #위도/경도
                        radius=df_2013_gu.loc[item,'전기']/100,
                        popup=df_2013_gu.loc[item,'전기'],
                        color ='crimson',
                        fill = True).add_to(map_osm)
    folium.CircleMarker([lat,long],  #위도/경도
                    radius=df_2013_gu.loc[item,'가스']/100,
                    popup=df_2013_gu.loc[item,'가스'],
                    color ='blue',
                    fill = True).add_to(map_osm)
    folium.CircleMarker([lat,long],  #위도/경도
                    radius=df_2013_gu.loc[item,'지역난방']/100,
                    popup=df_2013_gu.loc[item,'지역난방'],
                    color ='green',
                    fill = True).add_to(map_osm)
map_osm
```

```python
#용도별  배출량원단위(전기)	배출량원단위(가스)	배출량원단위(지역난방)

#group by 
dfg2=data.groupby(['연도','용도명'])['전기','가스','지역난방'].mean()
dfg2
```

```python
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc
font_name = font_manager.FontProperties(fname="c:/Windows/Fonts/malgun.ttf").get_name()
rc('font', family=font_name)

dfg2.plot(kind='bar',figsize=(10,6))
plt.show()
```
