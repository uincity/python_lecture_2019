```python
#plot 크기 지정 
df1.plot(figsize=(10,6))
plt.show()
```

```python
## bar그래프
df1.plot(kind='bar',figsize=(10,6))
plt.show()
```

```python
## 음식물 발생량으로만 그래프 그리기,발생량 많은순서로
df1['음식물류발생량'].sort_values().plot(kind='barh',color='blue',figsize=(10,6))
plt.show()
```
