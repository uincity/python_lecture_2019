#### 4.1. 부산시기온 읽어와서 리스트로 처리하기

```python
f = open('./data/부산시기온.csv')
data = f.readlines()
f.close

print(data)
```

```python
#일시/평균/최고/최저 4가지 항목의 리스트 생성
day = []
tempavg = []
tempmax = []
tempmin = []
```

```python
del data[0]
print(data)
```

```python
for line in data:
    line = line.replace('\n','')   # return 문자 공백처리
    item = line.split(',')
    day.append(item[0])
    tempavg.append(float(item[1]))
    tempmax.append(float(item[2]))
    tempmin.append(float(item[3]))

print(day)
print(tempavg)
print(tempmax)
print(tempmin)
```

```python
# dictionary  일평균에 대한 평균값 딕셔너리 만들기
day_avg = dict(zip(day,tempavg))
type(day_avg)
print(day_avg)
```

```python
import matplotlib.pyplot as plt 
from matplotlib import font_manager, rc
font_name = font_manager.FontProperties(fname="c:/Windows/Fonts/malgun.ttf").get_name()
rc('font', family=font_name)

x = range(len(tempmax))

plt.figure(figsize=(20, 8)) # 차트 사이즈 지정
plt.plot(tempmin, label='평균최저기온', marker='*')
plt.plot(tempmax, label='평균최고기온', marker='o' )
plt.plot(tempavg, label='평균기온', marker='^' )

plt.xlabel('연도')
plt.ylabel('기온')
plt.title('부산지역 연도별 평균기온 추이(1905~2018)')
plt.xticks(x, day,rotation=70)
plt.legend()
plt.show()
```
