#### 기상청 데이터 시각화 실습
#### 연도별 평균최저기온, 평균 최고기온 추이 시각화

```python
f= open('./data/201982117.csv','r')
data = f.readlines()
f.close

print(data)
```

```python
del data[0]
print(data)
```

```python
xyear = []
y1 = []
y2 = []
y3 = []

for line in data:
    line = line.replace('\n','')   # return 문자 공백처리
    item = line.split(',')
    xyear.append(item[0])
    y1.append(float(item[1]))
    y2.append(float(item[2]))
    y3.append(float(item[3]))

print(xyear)
print(y1)
print(y2)
```

```python
import matplotlib.pyplot as plt 
from matplotlib import font_manager, rc
font_name = font_manager.FontProperties(fname="c:/Windows/Fonts/malgun.ttf").get_name()
rc('font', family=font_name)

x = range(len(y2))

plt.figure(figsize=(20, 8)) # 차트 사이즈 지정
plt.plot(y1, label='평균최저기온', marker='*')
plt.plot(y2, label='평균최고기온', marker='o' )
plt.xlabel('연도')
plt.ylabel('기온')
plt.title('부산지역 연도별 평균기온 추이(1905~2018)')
plt.xticks(x, xyear,rotation=70)
plt.legend()
plt.show()
```
